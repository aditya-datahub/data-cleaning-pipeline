# Handling missing data

```python
df.dropna(subset = ["Customer_ID"], inplace = True)
```

```python
df.isnull().sum()
```

```python
df.head(2)
```

```python
df['Age'].unique()
```

```python
import re

def extract_age(age):
    age_num =re.findall('[0-9]+', str(age))
    if len(age_num)>0:
        return age_num[0]
    else:
        return age

df["Age"] = df["Age"].apply(lambda x: extract_age(x))
```

```python
df_age = df[df['Age'] !=  'nan years']['Age'] 
```

```python
age_median = int(df_age.dropna().astype('int64').median())
```

```python
age_median
```

```python
df.replace('nan years', age_median, inplace = True)
```

```python
df['Age'].fillna(age_median, inplace = True) 
```

```python
df['Age'].unique()
```

```python
df['Purchase_Amount'].fillna(df['Purchase_Amount'].median(), inplace = True)
```

```python
df.isnull().sum()
```

```python
df['Feedback_Score'].fillna(df['Feedback_Score'].mode()[0], inplace = True)
```

```python
df.isnull().sum()
```

```python
for col in ['Gender', 'City', 'Country']:
    df[col].fillna(df[col].mode()[0], inplace= True)
```

```python
df['Last_Purchase_Date'].ffill(inplace = True)
```

```python
# Import necessary libraries
from scipy import stats
import numpy as np
import pandas as pd

# Step 1: Calculate Z-scores for 'Age' and 'Purchase_Amount'
z_scores = np.abs(stats.zscore(df[['Age', 'Purchase_Amount']]))

# Step 2: Identify rows with outliers (Z-score > 3)
outliers = (z_scores > 3).any(axis=1)
print(outliers)

# Step 3: Remove outliers and create a cleaned DataFrame
df_cleaned = df[(z_scores < 3).all(axis=1)]

# Step 4: Check the shape before and after removal
print("Original shape:", df.shape)
print("After removing outliers:", df_cleaned.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cols = ['Age', 'Purchase_Amount', 'Feedback_Score']

plt.figure(figsize=(15, 5))

for i, col in enumerate(cols, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(y=df_cleaned[col], color='skyblue', width=0.3)
    plt.title(f'Boxplot of {col}', fontsize=12)
    plt.ylabel('')
    plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
